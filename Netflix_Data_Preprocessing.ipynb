```python
import pandas as pd # Importing the Pandas library for data analysis and cleaning.

df = pd.read_csv("netflix_titles.csv") # Loading the dataset from the CSV file into a DataFrame.
df.head() # Displaying the first 5 rows of the dataset.
```

```python
# Showing information about the dataset.
# Includes column names, data types, and non-null values.
df.info()
```

```python
df.isnull().sum() # Checking for missing (null) values in each column
```

```python
df.fillna("Unknown", inplace=True) # Filling missing values with the word "Unknown".
```

```python
df.isnull().sum() # Re-checking for Null Values.
```

```python
df.duplicated().sum() # Checking the number of duplicate rows in the dataset.
```

```python
df.drop_duplicates(inplace=True) # Removing duplicate rows from the dataset.
```

```python
df.duplicated().sum() # Re-checking the number of duplicate rows in the dataset.
```

```python
# Converting all column names to lowercase.
# Replacing spaces with underscores for consistency.

df.columns = df.columns.str.lower().str.replace(" ", "_")
df.head()
```

```python
# Converting the 'date_added' column into datetime format.
# Invalid dates will be converted to NaT.

df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df.head() # Displaying the cleaned dataset.
```

```python
df.info()
```

```python
df.to_csv("cleaned_netflix_titles.csv", index=False) # Saving the cleaned dataset as a new CSV file.

from google.colab import files # Importing files module from Google Colab for downloading files.

files.download("cleaned_netflix_titles.csv") # Downloading the cleaned CSV file to the local computer.
```
